## Create a model

```
from PIL import Image
import fitz  # PyMuPDF，用于处理PDF
from google.colab import userdata
# import genai
import os
import time

def extract_text_from_image(image):
    """
    Extracts text from the given image using the genai model.

    Parameters:
    - image (PIL.Image): The image object.

    Returns:
    - str: The extracted text from the image.
    """
    # 获取Google API密钥
    GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')

    # 配置genai模型
    genai.configure(api_key=GOOGLE_API_KEY)
    model = genai.GenerativeModel(
        model_name='gemini-1.5-flash-002'
    )

    # 使用模型生成内容
    response = model.generate_content([
        image,
        ("给我图片中的原文")
    ])

    # 返回提取的文本内容
    return response.text

def pdf_to_text(pdf_path, output_txt_path, start_page=0):
    """
    Converts each page of a PDF to an image, extracts text, and saves it to a TXT file.

    Parameters:
    - pdf_path (str): The path to the PDF file.
    - output_txt_path (str): The path to save the extracted text file.
    - start_page (int): The page number to start processing from (0-indexed).
    """
    # 打开PDF文件
    pdf_document = fitz.open(pdf_path)
    all_text = ""

    # 逐页处理PDF文件，从start_page开始
    for page_num in range(start_page, pdf_document.page_count):

        # 将每一页转换为图像
        page = pdf_document.load_page(page_num)
        pix = page.get_pixmap()

        # 创建临时图像文件
        image_path = f"/tmp/page_{page_num}.png"
        pix.save(image_path)

        # 打开图像并提取文本
        im = Image.open(image_path)
        page_text = extract_text_from_image(im)
        print(page_text)
        all_text += f"\n--- Page {page_num + 1} ---\n{page_text}\n"

        # 删除临时图像文件
        os.remove(image_path)
        print(f"Processed page {page_num + 1}/{pdf_document.page_count}")

        # 等待3秒，避免触发API限流
        time.sleep(3)

    # 将所有提取的文本写入输出文件
    with open(output_txt_path, "w", encoding="utf-8") as f:
        f.write(all_text)

    print(f"Text extraction complete. Saved to {output_txt_path}")

# Example usage:
pdf_path = "/content/drive/My Drive/book all/jxdl.pdf"  # 你的PDF文件在Google Drive中的路径
output_txt_path = '/content/drive/My Drive/book all/jxdl.txt'

# 从第10页开始处理（注意：start_page是0-indexed，因此第10页传入9）
pdf_to_text(pdf_path, output_txt_path, start_page=235)
```
